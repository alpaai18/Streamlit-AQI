### Cleaning Data

**Planning:**
- Pembersihan null values menggunakan median
- Penanganan outlier (untuk meningkatkan kualitas data)
- Perubahan tipe data untuk kolom year, month, dan day menjadi datetime

```python
df_tiantan.info()
```

```python
for dataset in list_datasets:
  for column in list_columns:
    nilai_median = dataset[column].median()
    dataset.fillna(value=nilai_median, inplace=True)
```

```python
list_datasets = [df_aotizhongxin, df_changping, df_dingling, df_dongsi, df_guanyuan, df_gucheng, df_huairou, df_nongzhanguan, df_shunyi, df_tiantan, df_wanliu, df_wanshouxigong]
name_stations = ['aotizhongxin', 'changping', 'dingling', 'dongsi', 'guanyuan', 'gucheng', 'huairou', 'nongzhanguan', 'shunyi', 'tiantan', 'wanliu', 'wanshouxigong']
list_columns = ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3', 'TEMP', 'PRES', 'DEWP', 'WSPM']

for dataset in list_datasets:
  for column in list_columns:
    Q1 = np.percentile(dataset[column], 25)
    Q3 = np.percentile(dataset[column], 75)

    IQR = Q3 - Q1

    batas_atas = Q3 + (1.5*IQR)
    batas_bawah = Q1 - (1.5*IQR)

    min_outlier = dataset[column] < batas_bawah
    max_outlier = dataset[column] > batas_atas

    dataset[column].mask(cond=min_outlier, other=batas_bawah, inplace=True)
    dataset[column].mask(cond=max_outlier, other=batas_atas, inplace=True)
```

```python
def groupby_year(dataset):
  dataset = dataset.groupby(by='year').agg({
      "PM2.5": "mean",
      "PM10" : "mean",
      "SO2"  : "mean",
      "NO2"  : "mean",
      "CO"   : "mean",
      "O3"   : "mean",
      "TEMP" : "mean",
      "PRES" : "mean",
      "DEWP" : "mean",
      "RAIN" : "mean",
      "WSPM" : "sum",
      "station" : "first",
  }).reset_index()

  return dataset
```

```python
df_all['CO'] = df_all['CO']/1145
```

```python
# RUMUS PENENTUAN NILAI AQI

PM25_AQI = []
PM10_AQI = []
SO2_AQI = []
NO2_AQI = []
CO_AQI = []
O3_AQI = []

for index, kolom in df_all.iterrows():
# ====  PM2.5  ========================================================== #
  if 0 <= kolom['PM2.5'] < 12.1:
    PM25_AQI.append((kolom['PM2.5']-0)/(12.0-0)*(50-0)+0)

  elif 12.1 <= kolom['PM2.5'] < 35.5:
    PM25_AQI.append((kolom['PM2.5']-12.1)/(35.4-12.1)*(100-51)+51)

  elif 35.5 <= kolom['PM2.5'] < 55.5:
    PM25_AQI.append((kolom['PM2.5']-35.5)/(55.4-35.5)*(150-101)+101)

  elif 55.5 <= kolom['PM2.5'] < 150.5:
    PM25_AQI.append((kolom['PM2.5']-55.5)/(150.4-55.5)*(200-151)+151)

  elif 150.5 <= kolom['PM2.5'] < 250.5:
    PM25_AQI.append((kolom['PM2.5']-150.5)/(250.4-150.5)*(300-201)+201)

  elif 250.5 <= kolom['PM2.5'] < 500.5:
    PM25_AQI.append((kolom['PM2.5']-250.5)/(500.4-250.5)*(500-301)+301)

# ====  PM10  ========================================================== #

  if 0 <= kolom['PM10'] < 55:
    PM10_AQI.append((kolom['PM10']-0)/(54.9-0)*(50-0)+0)

  elif 55 <= kolom['PM10'] < 155:
    PM10_AQI.append((kolom['PM10']-55)/(154.9-55)*(100-51)+51)

  elif 155 <= kolom['PM10'] < 255:
    PM10_AQI.append((kolom['PM10']-155)/(254.9-155)*(150-101)+101)

  elif 255 <= kolom['PM10'] < 355:
    PM10_AQI.append((kolom['PM10']-255)/(354.9-255)*(200-151)+151)

  elif 355 <= kolom['PM10'] < 425:
    PM10_AQI.append((kolom['PM10']-355)/(424.9-355)*(300-201)+201)

  elif 425 <= kolom['PM10'] < 605:
    PM10_AQI.append((kolom['PM10']-425)/(604.9-425)*(500-301)+301)

# ====  SO2  ========================================================== #

  if 0 <= kolom['SO2'] < 36:
    SO2_AQI.append((kolom['SO2']-0)/(35-0)*(50-0)+0)

  elif 36 <= kolom['SO2'] < 76:
    SO2_AQI.append((kolom['SO2']-36)/(75-36)*(100-51)+51)

  elif 76 <= kolom['SO2'] < 186:
    SO2_AQI.append((kolom['SO2']-76)/(185-76)*(150-101)+101)

  elif 186 <= kolom['SO2'] < 305:
    SO2_AQI.append((kolom['SO2']-186)/(304-186)*(200-151)+151)

  elif 305 <= kolom['SO2'] < 605:
    SO2_AQI.append((kolom['SO2']-305)/(604-305)*(300-201)+201)

# ====  NO2  ========================================================== #

  if 0 <= kolom['NO2'] < 54:
    NO2_AQI.append((kolom['NO2']-0)/(53-0)*(50-0)+0)

  elif 54 <= kolom['NO2'] < 101:
    NO2_AQI.append((kolom['NO2']-54)/(100-54)*(100-51)+51)

  elif 101 <= kolom['NO2'] < 361:
    NO2_AQI.append((kolom['NO2']-101)/(360-101)*(150-101)+101)

  elif 361 <= kolom['NO2'] < 650:
    NO2_AQI.append((kolom['NO2']-361)/(649-361)*(200-151)+151)

  elif 650 <= kolom['NO2'] < 1250:
    NO2_AQI.append((kolom['NO2']-650)/(1249-650)*(300-201)+201)

# ====  CO  ========================================================== #

  if 0 <= kolom['CO'] < 4.5:
    CO_AQI.append((kolom['CO']-0)/(4.4-0)*(50-0)+0)

  elif 4.5 <= kolom['CO'] < 9.5:
    CO_AQI.append((kolom['CO']-4.5)/(9.4-4.5)*(100-51)+51)

  elif 9.5 <= kolom['CO'] < 12.5:
    CO_AQI.append((kolom['CO']-9.5)/(12.4-9.5)*(150-101)+101)

  elif 12.5 <= kolom['CO'] < 15.5:
    CO_AQI.append((kolom['CO']-12.5)/(15.4-12.4)*(200-151)+151)

  elif 15.5 <= kolom['CO'] < 30.5:
    CO_AQI.append((kolom['CO']-15.5)/(30.4-15.5)*(300-201)+201)

# ====  O3  ========================================================== #

  if 0 <= kolom['O3'] < 55:
    O3_AQI.append((kolom['O3']-0)/(54-0)*(50-0)+0)

  elif 55 <= kolom['O3'] < 71:
    O3_AQI.append((kolom['O3']-55)/(70-55)*(100-51)+51)

  elif 71 <= kolom['O3'] < 86:
    O3_AQI.append((kolom['O3']-71)/(85-71)*(150-101)+101)

  elif 86 <= kolom['O3'] < 106:
    O3_AQI.append((kolom['O3']-86)/(105-86)*(200-151)+151)

  elif 106 <= kolom['O3'] < 200:
    O3_AQI.append((kolom['O3']-106)/(200-106)*(300-201)+201)

df_all.insert(2, 'PM2.5_AQI', PM25_AQI)
df_all.insert(4, 'PM10_AQI', PM10_AQI)
df_all.insert(6, 'SO2_AQI', SO2_AQI)
df_all.insert(8, 'NO2_AQI', NO2_AQI)
df_all.insert(10, 'CO_AQI', CO_AQI)
df_all.insert(12, 'O3_AQI', O3_AQI)
```

```python
AQI = []

df_all_AQI = df_all[['PM2.5_AQI', 'PM10_AQI', 'SO2_AQI', 'NO2_AQI', 'CO_AQI', 'O3_AQI']]
for index, baris in df_all_AQI.iterrows():
  AQI.append(baris.max())

df_all.insert(13, 'AQI', AQI)
```

Dari visualisasi heatmap dapat diambil kesimpulan jika hubungan antara 2 variabel bernilai negatif maka akan memiliki hubungan tolak belakang (Jika Variabel A naik maka Variabel B akan turun). Dan sebaliknya jika positif maka akan memiliki hubungan sejalan (Variabel A naik maka variabel B naik)

**CONCLUSION PERTANYAAN 2**
- Dapat dilihat juga bahwa semua stasiun memiliki permasalahan pada polutan PM2.5, untuk penduduk setempat alangkah baiknya jika menggunakan masker saat beraktivitas khususnya untuk penderita asma ataupun penderita penyakit pernapasan lainnya.

- Untuk pemerintah setempat bisa melakukan tindakan yang dapat meredam konsentrasi polutan PM2.5 pada setiap stasiun dengan cara:
    - Melakukan penghijauan
    - Mengganti kendaraan dengan menggunakan energi ramah lingkungan
    - Pengurangan limbah
    - dan hal hal yang bisa meredam tekanan udara di daerah setempat
